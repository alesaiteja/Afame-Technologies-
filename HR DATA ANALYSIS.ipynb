```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\alesa\OneDrive\Desktop\Afame tech\HR Data.csv')
```

```python
# Display the first few rows 
df.head()
```

```python
df.info()
```

```python
# 1. Remove Unnecessary Columns
unnecessary_columns = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']  
df.drop(columns=unnecessary_columns, inplace=True)
```

```python
# 2. Rename Columns
df.rename(columns={'DistanceFromHome': 'Distance'}, inplace=True)
```

```python
# 3. Remove Redundant Entries
df.drop_duplicates(inplace=True)
```

```python
#Remove any null values and duplicates
df.dropna(inplace=True)
df.drop_duplicates()
```

```python
# 4. Eliminate NaN Values(Not a Number)
df.dropna(inplace=True)
```
